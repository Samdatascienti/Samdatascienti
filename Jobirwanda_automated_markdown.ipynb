### 0. INTRODUCTION <br>

The use of technology has steadily grown in Rwanda and the industry of employment was not missed out where a number of organizations using different websites to announce job offers. Those platforms serve as web platforms to post jobs such as Job in Rwanda, E-recruitment of mifotra, ndangira, LinkedIn and others. These websites have data related to employment such as number of searches and job applications through the platforms, domain of job, skills and so on. <br>

Mostly jobs posted on the mentioned websites are for educated people who have graduated with high school education and above. The public jobs with contract in Rwanda are advertisement on Mifotra websites E-recruitment, most of the jobs with contract of NGOs institutions are advertised in different job site recruitment engines. Private companies have different way of recruitment some use their own website and others use job site recruitment engines such as Job in Rwanda, umucuruzi, ndangira, LinkedIn and others. <br>

This data could complement the National Institute of Statistics of Rwanda through the labor force survey on employment related data. It can serve as a comparison of labor market supply based on jobs advertised, corresponding institutions and labor market demand based on jobs demanded with corresponding applicants. <br>

This report provides employment statistics obtained through web scraping the website of Job in Rwanda where 245 jobs were announced in the period of 10th January to 28th February 2022. Following statistics were computed such as job type, required skills, required experience, and degree of education, number of job positions by month, number of job adverts, and number of views for each job.  The descriptive statistics were calculated only on vacancies that are jobs and internship.

# 7. Correlation Analysis

Correlation between education level and years of experience required.

|Tests	|Values|
|------:|-----:|
|Chi-square	|69.13|
|P-value	|5.17284132400234e-06|


Based on values of p-value equal to 5.17284132400234e-06 which is less than 0.05. We reject null hypothesis stating that there is no relationship between education level and years of experience required. Therefore, the relation between education level and years of experience required is statistically significant.
