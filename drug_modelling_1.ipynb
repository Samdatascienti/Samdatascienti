```python
#Splitting data into independent and dependent variable
def preparation(dd,n_features):
  X,Y = [],[]
  for i in range(len(dd)):  
#determine end point
     end_point = i + n_features
#check if we are beyond the sequence
     if end_point>len(dd)-1:
       break
#gather input and output parts of this pattern
     seq_X,seq_Y=dd[i:end_point],dd[end_point]
     X.append(seq_X)
     Y.append(seq_Y)
  return np.array(X), np.array(Y)
```
