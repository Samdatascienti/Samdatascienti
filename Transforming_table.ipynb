# Labor Force Table 1

```python
df2=pd.read_excel('/content/gdrive/MyDrive/Big data and data revolution/RLFS Tables_ Annual_2020.xls',
                  sheet_name='Table 1', skiprows=2)
df2.columns=['indicators','total','Male','Female','Urban','Rural','PA','PN']
df2=df2.dropna(axis=0, how='all')
```

```python
df2.head()
```

```python
df_l=df2.melt(id_vars=['indicators'], var_name='charact', value_name='Value')
# df_l.charact.unique()
df_l.loc[df_l['charact'].isin(['total']), 'group']= 'total'
df_l.loc[df_l['charact'].isin(['Male', 'Female']), 'group']= 'Sex'
df_l.loc[df_l['charact'].isin(['Urban', 'Rural']), 'group']= 'Residence'
df_l.loc[df_l['charact'].isin(['PA', 'PN']), 'group']= 'Participated'
df_l.head()
```

```python
df_l.loc[df_l['indicators'].str.endswith('(%)'), 'Percent']= df_l[df_l['indicators'].str.endswith('(%)')]['Value']
df_l.loc[df_l['indicators'].str.endswith('(%)'), 'Value']=np.nan
df_l['indicators']=df_l['indicators'].str.replace('\-','')
df_l.head()
```

```python
df_l['indicators']=df_l['indicators'].str.strip()
df_l.head()
```

```python
df_l.loc[df_l['indicators'].str.endswith('job'), 'median_monthly']= df_l[df_l['indicators'].str.endswith('job')]['Value']
df_l.loc[df_l['indicators'].str.endswith('job'), 'Value']=np.nan
df_l.head()
```

```python
# 
df_l['Sex']=np.nan
df_l.Sex[df_l.group == 'Sex'] = df_l.charact
df_l['Residence']=np.nan
df_l.Residence[df_l.group == 'Residence'] = df_l.charact
df_l['Participated']=np.nan
df_l.Participated[df_l.group == 'Participated'] = df_l.charact
df_l.head()
```

```python
# # creating new column of indicators related with percentage
df_l.loc[df_l['indicators'].str.endswith('(%)'), 'Indicators_per']= df_l[df_l['indicators'].str.endswith('(%)')]['indicators']
df_l.loc[df_l['indicators'].str.endswith('(%)'), 'indicators']=''
# creating new column of indicators related with money
df_l.loc[df_l['indicators'].str.endswith('job'), 'Indicators_inc']= df_l[df_l['indicators'].str.endswith('job')]['indicators']
df_l.loc[df_l['indicators'].str.endswith('job'), 'indicators']=''
df_l
```

```python
df_l[df_l['Indicators_per']=='LU1  Unemployment rate(%)']
```

```python
df_l['Indicators_per']=df_l['Indicators_per'].str.replace(" ","")
df_l
```

```python

options=['Labour force participation rate(%)','Employmenttopopulation ratio(%)','LU1  Unemployment rate(%)',
         'Youth unemployment rate (1630 yrs)(%)']

df_lclean=df_l[df_l['Indicators_per'].isin(options)]
df_lclean=df_lclean.drop(['indicators','charact',	'group','Value','median_monthly',	'Participated', 'Indicators_inc'], axis = 1)
df_lclean=df_lclean.drop([7,8,9,10,67,68,69,70,79,80,81,82])
df_lclean
```

```python
# df_lclean['Indicators_per']=df_lclean['Indicators_per'].str.replace(" ","")
# df_lclean
```

```python
# Export to csv
df_lclean.to_csv('RLFS_table1.csv')
files.download('RLFS_table1.csv')
```
